### Data Preparation

```python
# Menyiapkan data dan modul yang dibutuhkan
import pandas as pd
import seaborn as sns

df_supermarket = pd.read_csv('EDA-supermarket.csv')
```

```python
# Melihat beberapa sampel data
df_supermarket.sample(10)
```

```python
# Memeriksa kesesuaian tipe data atribut
df_supermarket.dtypes
```

```python
# Berdasarkan informasi di atas, terdapat beberapa atribut yang memiliki tipe data yang kurang tepat.
df_supermarket['Order Date'] = pd.to_datetime(df_supermarket['Order Date'], dayfirst=True)
df_supermarket['Ship Date'] = pd.to_datetime(df_supermarket['Ship Date'], dayfirst=True)
df_supermarket['Ship Mode'] = df_supermarket['Ship Mode'].astype('category')
df_supermarket['Segment'] = df_supermarket['Segment'].astype('category')
df_supermarket['Category'] = df_supermarket['Category'].astype('category')
df_supermarket['Sub-Category'] = df_supermarket['Sub-Category'].astype('category')
```

```python
# Memeriksa data yang hilang
df_supermarket.isnull().sum()
```

```python
# Berdasarkan informasi di atas, diperoleh bahwa terdapat nilai yang hilang pada atribut "Postal Code".
df_supermarket = df_supermarket.dropna()
```

```python
# Menambahkan atribut "Order Date Name" yang merupakan nama hari dari "Order Date"
df_supermarket['Order Date Name'] = df_supermarket['Order Date'].dt.day_name()
```
